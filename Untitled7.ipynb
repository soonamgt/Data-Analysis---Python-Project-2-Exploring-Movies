# Now we will begin our Data Visualization 

Q.1- What is the distribution of movie ratings in our dataset?

```python
#Answer: We will use a histogram to showcase the distribution of ratings, helping us identify the most common rating ranges.
```

```python
sns.histplot(df['rating'], bins=20, kde=True, color='pink')
sns.set(style="whitegrid")
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.figure(figsize=(10, 6))
plt.show()
```

Q.2- How does the gross earnings vary across different movie genres?

```python
#Answer: A grouped bar chart can illustrate the variation in gross earnings for each genre, highlighting 
#which genres tend to have higher or lower earnings.
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='genre', y='gross', data=df, palette='viridis')
plt.title('Gross Earnings Across Different Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Gross Earnings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Q.3- Is there a correlation between movie budget and gross earnings?

```python
#Answer:To analyze the correlation between movie budget and gross earnings,we will use a scatter plot with a regression line. 
```

```python
plt.figure(figsize=(12, 8))
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Correlation Between Movie Budget and Gross Earnings')
plt.xlabel('Budget (in millions)')
plt.ylabel('Gross Earnings (in millions)')
plt.grid(True)
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap with seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

# Add a title
plt.title('Correlation Matrix')

# Display the plot
plt.show()
```

```python
#Positive Slope: If the regression line slopes upward from left to right, it suggests a positive correlation. This means that, 
#on average, movies with higher budgets tend to have higher gross earnings. As per the analysis through correlation matrix , 
#its 74% between movies budget and gross earnings.
```

Q.4- Question: Which countries produce the highest number of movies?

```python
#Answer: A bar chart or a world map can help visualize the production volume across different countries, identifying the 
#top contributors.
```

```python
movies_per_country = df['country'].value_counts().reset_index()
movies_per_country.columns = ['country', 'movie_count']

# Plotting the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='movie_count', y='country', data=movies_per_country.head(10), palette='muted')
plt.title('Top 10 Countries Producing the Highest Number of Movies')
plt.xlabel('Number of Movies Produced')
plt.ylabel('Country')
plt.show()
```

Q.5- : How does the performance of the top production companies compare in terms of gross earnings?

```python
#Certainly! To compare the performance of the top production companies in terms of gross earnings, we can create a pie chart
```

```python
import matplotlib.pyplot as plt

# Group by production companies and sum their gross earnings
company_gross = df.groupby('company')['gross'].sum()

# Select the top N production companies (adjust N as needed)
top_companies = company_gross.nlargest(10)

# Plot a pie chart
plt.figure(figsize=(10, 10))
top_companies.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Top Production Companies by Gross Earnings')
plt.ylabel('')
plt.show()
```

Q.6- Question: Are there any outliers in the distribution of movie votes?

Q.7- How are movie ratings distributed within each genre?

```python
df= df.dropna(subset=['rating', 'genre'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your cleaned DataFrame after dropping missing values
if df.empty:
    print("Warning: The DataFrame is empty. Please inspect the data.")
else:
    plt.figure(figsize=(14, 8))
    sns.countplot(x='genre', data=df, color='red')
    plt.title('Distribution of Movie Ratings Within Each Genre')
    plt.xlabel('Genre')
    plt.ylabel('Number of Movies')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```
